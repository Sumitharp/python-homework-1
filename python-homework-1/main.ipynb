```python
#main file


# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path('./Resources/budget_data.csv')


# Initialize list of records
records = []

# Open the csv file as an object
with open(csvpath, 'r') as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    
    csvreader = csv.reader(csvfile, delimiter=',')

    # Read the header row first (skip this step if there is no header)
    
    csv_header = next(csvreader)
    
    # Added Change_pnl for recording Change profit and loss
    csv_header.append("Change_PnL")
    
    
    records.append(csv_header)
    
    
    # Initialise variables
    count = 0
    total_pnl=0
    change_pnl=0
    pre_pnl=0 # profit and loss for previous period to calculate change
    sum_change_pnl=0
    max_change_pnl=0
    min_change_pnl=float('inf')
  
    # Read each row of data and calculate 1) pre_pnl to calculate change_pnl then add change_ppl values under the new header column 
    # 2) calculate max change and min change values and also count no of months. 
    # 3) Next add monthly profit and loss to total_pnl for each row
    
    for row in csvreader:
        if pre_pnl==0:    # only for the first period
            row.append(int(0))
        else:  
            change_pnl=int(row[1])-pre_pnl
            sum_change_pnl+=change_pnl
            row.append(change_pnl)
            
        
        pre_pnl=int(row[1])
         
        total_pnl+=int(row[1])
        
        records.append(row)
        
        
        if change_pnl > max_change_pnl:
            max_change_pnl=change_pnl
            max_change_month=row[0]
            
        if (change_pnl<min_change_pnl):
            min_change_pnl=change_pnl
            min_change_month=row[0]
            
        count+=1
        
    avg_change=round(sum_change_pnl/(count-1),2)
    
    print(f" Financial Analysis")
    print("--------------------")
    print(f"Total no of months:{count}")
    print(f"Total profit and loss :${total_pnl}")
    print(f"Average of Change:${avg_change}")
    print(f"Greatest Increase in Profits:{max_change_month} (${max_change_pnl})")
    print(f"Greatest Decrease in Profits:{min_change_month} (${min_change_pnl})")
    
# Set the path for the output.csv
x = Path("output_budget_pybank.csv")

# Open the output path as a file and pass into the 'csv.writer()' function
# Set the delimiter/separater as a ','
with open(x, 'w') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=',')

    # Loop through the list of records and write every record to the
    # output csv file
    for record in records:
        csvwriter.writerow(record)
```
